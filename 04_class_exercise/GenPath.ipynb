```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.path as path
```

### First General Path Example


```python
vertices = [
    (50, 50),
    (50, 200),
    (150, 500), (250, 200),
    (350, 100), (150, 150), (100, 100),
    (50, 50)
]

codes = [
    path.Path.MOVETO,
    path.Path.LINETO,
    path.Path.CURVE3, path.Path.CURVE3,
    path.Path.CURVE4, path.Path.CURVE4, path.Path.CURVE4,
    path.Path.CLOSEPOLY,
]
```

```python
path1 = path.Path(vertices, codes)

patch1 = patches.PathPatch(
    path1, 
    facecolor='yellow', 
    edgecolor='blue',
    lw=2
)
```

```python
fig, ax = plt.subplots()
ax.add_patch(patch1)
ax.set_xlim(0, 500)
ax.set_ylim(0, 500)
plt.show()
```

### Second General Path Example


```python
vertices_car = [
    (60, 120),
    (80, 120),
    (90, 140), (100, 120),
    (160, 120), 
    (170, 140), (180, 120),
    (200, 120),
    (195, 100), (200, 80), (160, 80),
    (110, 80),
    (60, 100),
    (60, 120)
]

codes_car = [
    path.Path.MOVETO,
    path.Path.LINETO,
    path.Path.CURVE3, path.Path.CURVE3,
    path.Path.LINETO,
    path.Path.CURVE3, path.Path.CURVE3,
    path.Path.LINETO,
    path.Path.CURVE4, path.Path.CURVE4, path.Path.CURVE4,
    path.Path.LINETO,
    path.Path.LINETO,
    path.Path.CLOSEPOLY,
]
```

```python
path_car = path.Path(vertices_car, codes_car)

patch_car = patches.PathPatch(
    path_car, 
    facecolor='yellow', 
    edgecolor='blue',
    lw=2
)
```

```python
fig, ax = plt.subplots()
ax.add_patch(patch_car)
ax.set_xlim(0, 500)
ax.set_ylim(0, 500)
plt.show()
```
