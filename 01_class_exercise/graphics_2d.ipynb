### Quadratic Bezier Curve


The formula for a quadratic Bezier curve is

B(t) = ((1-t)^2)(P0) + 2(1-t)(t)(P1) + (t^2)(P2)


```python
def quad_curve(t, p0, p1, p2):
    def point(index):
        first_grp = ((1.0 - t) ** 2.0) * p0[index]
        second_grp = 2.0 * (1.0 - t) * t * p1[index]
        third_grp = (t ** 2.0) * p2[index]
        
        return first_grp + second_grp + third_grp
        
    x = point(0)
    y = point(1)

    return x, y
```

```python
p0 = (50, 300)
p1 = (250, 150)
p2 = (350, 200)
```

```python
quad_cur_img = Image.new('RGB', (500, 500), (10, 23, 9))
quad_cur_draw = ImageDraw.Draw(quad_cur_img)

points = [quad_curve(t, p0, p1, p2) for t in [i/100 for i in range(101)]]
quad_cur_draw.line(points, fill='brown', width=5)

quad_cur_img.save('graphics_2d/quad_cur_img.png')
```

```python
quad_cur_img
```

### Cubic Bezier Curve


```python
def cubic_curve(t, p0, p1, p2, p3):
    def point(index):
        first_grp = ((1.0 - t) ** 3.0) * p0[index]
        second_grp = 3.0 * ((1.0 - t) ** 2.0) * t * p1[index]
        third_grp = 3.0 * (1.0 - t) * (t ** 2.0) * p2[index]
        fourth_grp = (t ** 3) * p3[index]
        
        return first_grp + second_grp + third_grp + fourth_grp
        
    x = point(0)
    y = point(1)

    return x, y
```

```python
p0 = (50, 300)
p1 = (100, 375)
p2 = (350, 200)
p3 = (250, 150)
```

```python
cubic_cur_img = Image.new('RGB', (500, 500), (10, 23, 9))
cubic_cur_draw = ImageDraw.Draw(cubic_cur_img)

points = [cubic_curve(t, p0, p1, p2, p3) for t in [i/100 for i in range(101)]]
cubic_cur_draw.line(points, fill='brown', width=5)

cubic_cur_img.save('graphics_2d/cubic_cur_img.png')
```

```python
cubic_cur_img
```

```python
cubic_p0 = (300, 200)
cubic_p1 = (350, 385)
cubic_p2 = (400, 185)
cubic_p3 = (450, 350)
```
