```python
from PIL import Image, ImageDraw
```

```python
first_img = Image.new('RGB', (1000, 1000), (255, 255, 255))
first_draw = ImageDraw.Draw(first_img)
```

```python
first_draw.line((100, 250, 150, 250), fill='black', width=5)  #ab
first_draw.line((150, 250, 150, 200), fill='black', width=5)  #bc
first_draw.line((150, 200, 200, 200), fill='black', width=5)  #cd
first_draw.line((200, 200, 200, 500), fill='black', width=5)  #de
first_draw.line((200, 500, 400, 500), fill='black', width=5)  #ef
first_draw.line((400, 500, 400, 200), fill='black', width=5)  #fg
first_draw.line((400, 200, 450, 200), fill='black', width=5)  #gh
first_draw.line((450, 200, 450, 250), fill='black', width=5)  #hi
first_draw.line((450, 250, 500, 250), fill='black', width=5)  #ij
first_draw.line((500, 250, 500, 150), fill='black', width=5)  #jk
first_draw.line((500, 150, 450, 100), fill='black', width=5)  #kl
first_draw.line((450, 100, 150, 100), fill='black', width=5)  #lm 
first_draw.line((150, 100, 100, 150), fill='black', width=5)  #mn
first_draw.line((100, 150, 100, 250), fill='black', width=5)  #na

first_img
```
